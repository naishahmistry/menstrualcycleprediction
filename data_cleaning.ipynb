```python
# Import libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# Read in data
data = pd.read_csv('data.csv')
print(data.columns)
data.head()
```

```python
# Filter columns to features & target for Bayesian model
cols_to_keep = ['sedentary_mins', 'active_mins', 'glucose_mean', 'rhr_mean', 'minutes_asleep', 'minutes_awake',
'sleep_efficiency_mean', 'sleep_score_overall', 'sleep_score_revitalization', 'sleep_score_duration', 'lh', 'estrogen', 'phase']
features_target = data[cols_to_keep]
features_target.head()
```

```python
# Check nan values
features_target.isna().sum()
```

```python
# Drop all nan values
features_target = features_target.dropna()
features_target.isna().sum()
features_target.head()
```

```python
features_target.to_csv('cleaned_data.csv')
```
